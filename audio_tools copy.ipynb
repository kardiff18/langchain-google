```python
from langchain_core.pydantic_v1 import BaseModel, Field

class Joke(BaseModel):  # Uses langchain_core.pydantic_v1.BaseModel
    """Joke to tell user."""

    setup: str = Field(description="question to set up a joke")
    punchline: str = Field(description="answer to resolve the joke")

model = ChatVertexAI(model_name="gemini-1.0-pro-001", temperature=0)


# Pydantic class
chat = model.with_structured_output(Joke)
result = chat.invoke("Tell me a joke about cats.")
print(result)
print(isinstance(result, Joke))

for chunk in chat.stream("Tell me a joke about cats."):
    print(isinstance(chunk, Joke))

# Schema
chat2 = model.with_structured_output(Joke.schema())
result = chat2.invoke("Tell me a joke about cats.")
print(result)
print(isinstance(result, dict))
print(set(result.keys()) == {"setup", "punchline"})

for chunk in chat2.stream("Tell me a joke about cats."):
    print(isinstance(chunk, dict))
print(isinstance(chunk, dict))  # for mypy
print(set(chunk.keys()) == {"setup", "punchline"})
```
